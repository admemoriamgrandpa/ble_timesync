```python
#first implementation of zero crossing

import numpy as np
import matplotlib.pyplot as plt
```

```python
# zero crossing by definition
x = np.linspace(100, 10*np.pi, 1000) #sine function

sinewave = np.sin(x)

# Annotation
plt.plot(sinewave)
plt.axhline(0, color='green')
plt.show()
```

```python
sample_mean = sum(sinewave)/len(sinewave)
```

```python
compute_zero_crossing = []

#this is not the most optimal way I could check for both reverse signs
for i in range(len(sinewave)):
    if not i == (len(sinewave)-1):
        if ((sinewave[i]*1 < 0 and sinewave[i + 1] * 1 > 0)):
            compute_zero_crossing.append(sinewave[i])
            compute_zero_crossing.append(sinewave[i + 1])
```

```python
#proof of concept
testout = [i - sample_mean for i in sinewave]
```

```python
# Annotation
plt.plot(testout, color='b')
plt.plot(sinewave, color='r')
plt.axhline(0, color='green')
plt.show()
```

```python
# Testing proof of concept with noise
buks = np.linspace(100, 10*np.pi, 1000) #sine function

sin_wav = np.sin(buks)
noise = np.random.normal(0, 1, buks.shape)

sin_wav = noise + sin_wav

# Annotation
plt.plot(sin_wav)
plt.axhline(0, color='green')
plt.show()
```

```python
sample_avg = sum(sin_wav)/len(sin_wav)
```

```python
#proof of concept
test = [i - sample_avg for i in sin_wav]
```

```python
# Annotation
plt.plot(test, color='b')
plt.plot(sin_wav, color='r')
plt.axhline(0, color='green')
plt.show()
```
